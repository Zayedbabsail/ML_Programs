```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import statsmodels.api as sm
from scipy import stats
```

```python
x=[1,2,3,4,5] 
y=[1,1,2,2,4]
```

```python
plt.scatter(x,y)
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

```python
x1 = sm.add_constant(x)
results = sm.OLS(y,x1).fit()
results.summary
slope=0.7
intercept=-0.1
```

```python
def myfun(x):
    return slope * x + intercept
mymodel = list(map(myfun,x))
mymodel
plt.scatter(x,y)
fig=plt.plot(x,mymodel,lw=4,c='orange',linestyle= '--')
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

```python
slope,intercept,r,p,std_err=stats.linregress(x,y)
print('\n',slope,'\n',intercept,'\n',r,'\n',p,'\n',std_err)
```

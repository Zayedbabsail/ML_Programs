```python
import matplotlib.pyplot as plt 
from scipy import stats
```

```python
x= [ 5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,80]
x1= [ 5,7,8,7,2,9,4,11,12,9,6]
y1=[99,86,87,88,103,87,94,78,77,85,80]
slope,intercept,r,p,std_err=stats.linregress(x,y)
print('with outliers\n',slope,'\n',intercept,'\n',r,'\n',p,'\n',std_err,'\n')
slope1,intercept1,r1,p1,std_err1=stats.linregress(x1,y1)
print('without outliers\n',slope1,'\n',intercept1,'\n',r1,'\n',p1,'\n',std_err1)
```

```python
def myfun(x):
    return slope * x + intercept
mymodel1 = list(map(myfun,x))


def myfun1(x1):
    return slope1 * x1 + intercept1
mymodel2=list(map(myfun1,x1))

print("R2",r*r)
print("After removing outliers R2", r1*r1)

plt.scatter(x,y)
plt.scatter(x1,y1)
fig=plt.plot(x,mymodel1,lw=4,c='orange',linestyle= '--',label='full')
fig1=plt.plot(x1,mymodel2,lw=4,c='red',linestyle= '-',label='outlier removed')
plt.xlabel("Years After purchase")
plt.ylabel("Resale value")
plt.legend()
```
